```julia
using Plots, Statistics, DelimitedFiles
n_sample = 50

times_nmon = zeros(n_sample,9)
times_nmon_small = zeros(n_sample,9)
times_pigfarm_singleV = zeros(n_sample,9)
times_pigfarm = zeros(n_sample,9)

for i in 1:n_sample
    result_nmonitoring = readdlm((@__DIR__)*"/results/nmonitoring_"*string(i)*".csv", ',')
    times_nmon[i,:] .= result_nmonitoring[2,2:end]
    result_nmonitoring_small = readdlm((@__DIR__)*"/results/nmonitoring_small_"*string(i)*".csv", ',')
    times_nmon_small[i,:] .= result_nmonitoring_small[2,2:end]
    result_pigfarm = readdlm((@__DIR__)*"/results/pigfarm_"*string(i)*".csv", ',')
    times_pigfarm[i,:] .= result_pigfarm[2,2:end]
    result_pigfarm_singleV = readdlm((@__DIR__)*"/results/pigfarm_singleV_"*string(i)*".csv", ',')
    times_pigfarm_singleV[i,:] .= result_pigfarm_singleV[2,2:end]
end



scatter(collect(2:10), mean(times_nmon, dims=1)', xlabel="Number of decision nodes", ylabel="Solution time (s)", label="N-monitoring", color=palette(:tab10)[1], legend=:topleft, yaxis=:log, yticks=[10.0^k for k in -3:3])
scatter!(collect(2:10), mean(times_pigfarm, dims=1)', label="Pig farm", color=palette(:tab10)[2])
scatter!(collect(2:10), mean(times_pigfarm_singleV, dims=1)', label="Pig farm, single value node", color=palette(:tab10)[3])
scatter!(collect(2:10), mean(times_nmon_small, dims=1)', label="N-monitoring, smaller RJT", color=palette(:tab10)[4])
plot!(collect(2:10), mean(times_nmon, dims=1)', label=false, ls=:dash, color=palette(:tab10)[1])
plot!(collect(2:10), mean(times_pigfarm, dims=1)', label=false, ls=:dash, color=palette(:tab10)[2])
plot!(collect(2:10), mean(times_pigfarm_singleV, dims=1)', label=false, ls=:dash, color=palette(:tab10)[3])
plot!(collect(2:10), mean(times_nmon_small, dims=1)', label=false, ls=:dash, color=palette(:tab10)[4])
# Plots.pdf("sol_times.pdf")
```
