```julia
using Plots, Statistics, DelimitedFiles
n_sample = 50

times_nmon = zeros(n_sample,10)
times_pigfarm_singleV = zeros(n_sample,10)
times_pigfarm = zeros(n_sample,10)

for i in 1:n_sample
    result_nmonitoring = readdlm((@__DIR__)*"/results/nmonitoring_"*string(i)*".csv", ',')
    times_nmon[i,:] .= result_nmonitoring[2,:]
    result_pigfarm = readdlm((@__DIR__)*"/results/pigfarm_"*string(i)*".csv", ',')
    times_pigfarm[i,:] .= result_pigfarm[2,:]
    result_pigfarm_singleV = readdlm((@__DIR__)*"/results/pigfarm_singleV_"*string(i)*".csv", ',')
    times_pigfarm_singleV[i,:] .= result_pigfarm_singleV[2,:]
end


scatter(collect(1:10), mean(times_nmon, dims=1)', xlabel="Number of decision nodes", ylabel="Solution time (s)", label="N-monitoring", color=palette(:tab10)[1], yaxis=:log)
scatter!(collect(1:10), mean(times_pigfarm, dims=1)', label="Pig farm", color=palette(:tab10)[2])
scatter!(collect(1:10), mean(times_pigfarm_singleV, dims=1)', label="Pig farm, single value node", color=palette(:tab10)[3])
plot!(collect(1:10), mean(times_nmon, dims=1)', label=false, ls=:dash, color=palette(:tab10)[1])
plot!(collect(1:10), mean(times_pigfarm, dims=1)', label=false, ls=:dash, color=palette(:tab10)[2])
plot!(collect(1:10), mean(times_pigfarm_singleV, dims=1)', label=false, ls=:dash, color=palette(:tab10)[3])
# Plots.pdf("sol_times.pdf")
```
